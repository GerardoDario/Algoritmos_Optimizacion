Problema 1. Organizar sesiones de doblaje(I)<br>
Se precisa coordinar el doblaje de una película. <br>
Los actores del doblaje deben coincidir en las tomas en las que sus personajes aparecen juntos en las diferentes tomas. <br>
Los actores de doblaje cobran todos la misma cantidad por cada día que deben desplazarse hasta el estudio de
grabación independientemente del número de tomas que se graben. <br>
No es posible grabar más de 6 tomas por día. <br>

El objetivo es planificar las sesiones por día de manera que el gasto por los
servicios de los actores de doblaje sea el menor posible. Los datos son:<br>

Número de actores: 10<br>
Número de tomas : 30<br>

- 1 indica que el actor participa en la toma
- 0 en caso contrario





                                        

# Modelo
- ¿Como represento el espacio de soluciones?
- ¿Cual es la función objetivo?
- ¿Como implemento las restricciones?

### Respuesta: <br><br>

$D = ${ $1,...,l$ }$ :$ como conjunto de días<br>
$A = ${ $1,...,m$ }$ :$ como conjunto de actores<br>
$T = ${ $1,...,n$ }$ :$ como conjunto de tomas<br>

**Parámetros:**
- $c$: Costo de cada actor por día
- max_takes: máximo de tomas que pueden ser grabadas por día
- $assignment_{at}$ : asignación de los actores a las tomas, es decir, dicho parámetro es igual a 1 si a un actor a perteneciente A se le asigna tomar t perteneciente T, en caso contrario, 0<br>


**Variables de decisión**<br><br>
$x_{td}$  $\in $ {0,1} : 1 si la toma $t $ $\in $ $ T $ es programada para el día $d $ $\in $ $ D $ , en otro caso: 0<br>
$y_{ad}$  $\in $ {0,1} : 1 si la toma $a $ $\in $ $ A $ es programada para el día $d $ $\in $ $ D $ , en otro caso: 0
<br><br>

**Modelo Matemático**<br><br>

- $min \sum \limits _{a \in A} \sum \limits _{d \in D} c $ $ y_{ad} $
- $\sum \limits _{d \in D} x_{td} = 1$, $t \in $ $T$
- $\sum \limits _{t \in T} x_{td} \leq max\_takes$, $d \in D$

- $y_{ad} \geq assignment_{at} \cdot x_{td}$, $a \in A$, $d \in D$, $t \in T$

- $x_{td} \in$ {0,1} , $t \in $ $T$ $d \in D$

# Análisis
- ¿Que complejidad tiene el problema?. Orden de complejidad y Contabilizar el espacio de soluciones

### Respuesta
Recorremos cada fila de la tabla, y para cada fila, recorremos cada columna (actor) y vemos si el actor está en la escena.<br>
Si el actor está en la escena, sumamos 1 a la cantidad de escenas en las que aparece el actor.<br>
Si el actor no está en la escena, no hacemos nada.<br>
Por lo tanto, la complejidad es O(np) donde n es la cantidad de escenas y p es la cantidad de actores<br>

# Diseño
- ¿Que técnica utilizo? ¿Por qué?

### Respuesta
Dada la dimensionalidad de los datos, utilizaré una técnica voraz para resolver el problema.<br>
Además, añado una poda heurística de entrada (actor con menos escenas) con la finalidad de reducir dimensiones en el espacio de solución.<br>
La técnica consiste en ir seleccionando el actor con menor cantidad de escenas en las que aparece.<br>
Luego eliminamos las escenas en las que aparece ese actor.<br>
De esta forma, reducimos la cantidad de escenas en las que aparece ese actor, y así sucesivamente con los demás actores.<br>

```python
# Librerias
import pandas as pd
import re
```

```python
def convert_google_sheet_url(url):
    # Regular expression to match and capture the necessary part of the URL
    pattern = r'https://docs\.google\.com/spreadsheets/d/([a-zA-Z0-9-_]+)(/edit#gid=(\d+)|/edit.*)?'

    # Replace function to construct the new URL for CSV export
    # If gid is present in the URL, it includes it in the export URL, otherwise, it's omitted
    replacement = lambda m: f'https://docs.google.com/spreadsheets/d/{m.group(1)}/export?' + (f'gid={m.group(3)}&' if m.group(3) else '') + 'format=csv'

    # Replace using regex
    new_url = re.sub(pattern, replacement, url)

    return new_url
```

```python
# Replace with your modified URL
url = 'https://docs.google.com/spreadsheets/d/1Ipn6IrbQP4ax8zOnivdBIw2lN0JISkJG4fXndYd27U0/edit#gid=0'

new_url = convert_google_sheet_url(url)

print(new_url)
# https://docs.google.com/spreadsheets/d/1mSEJtzy5L0nuIMRlY9rYdC5s899Ptu2gdMJcIalr5pg/export?gid=1606352415&format=csv
```

```python
# Importamos el archivo
df = pd.read_csv(new_url,header=1, index_col=0)
df = df.head(30)
df = df.iloc[:,:-2]
df
```

```python
# crea una funcion que calcula el maximo de cada columna y retorna la suma de los maximos
def maximo(df):
    # agrega una fila con el maximo de cada columna
    df.loc['max'] = df.max()
    # calcula la suma de las columnas 1,2,3,4,5,6,7,8,9 y 10 donde se encuentran los maximos
    df['suma_maximos'] = df.iloc[:,0:10].sum(axis=1)
    # retorna columna suma_maximos en index max
    return df['suma_maximos']['max']
```

```python
# crea una funcion que calcula el minimo de cada columna y retorna la suma de los minimos
def greedy_optimize(_data):

    days = []
    data = _data.copy()
    while len(data) > 0:
        costs_by_actor = data.sum()
        # Toma el actor con menor cantidad de escenas
        actor = costs_by_actor.idxmin()
        # Toma las escenas en las que aparece el actor
        scenes = data[data[actor] > 0].index
        # Obtiene los actores en las escenas del actor con menor cantidad de escenas
        actors = (data.loc[scenes].sum() > 0).astype(int)
        # Toma las escenas en las que aparecen los actores seleccionados
        scenes = ((data - actors) <= 0).all(axis=1)
        # Obtiene el indice de las primeras seis escenas
        scenes = scenes[scenes].index[:6]
        days.append(scenes)

        # Quita las escenas consideradas
        data = data.drop(scenes)
        data = data.drop(actor, axis=1)

    # Elimina los dias vacios
    days = [day for day in days if len(day) > 0]
    return days
```

```python
from termcolor import colored

days = greedy_optimize(df)
maximos = []
# para cada elemento de days, imprime el numero de dia y los actores en el dia
for i, day in enumerate(days):
    print(colored(f'Day {i+1}','red', attrs=['bold']))
    # imprime los actores con max igual a 1
    print(f'Actors: {", ".join(df.loc[day].max()[df.loc[day].max() == 1].index)}')
    # imprime las escenas en el dia
    print(f'Scenes: {", ".join(day)}')
    # imprime el costo del dia
    maximos.append(maximo(df.loc[day]))
    #print(f'Cost: {df.loc[day].sum().sum()}')
    print(f'Cost: {maximo(df.loc[day])}')
    # imprime una linea en blanco
    #print()


print("")
print(colored(f"Total cost: {sum(maximos)}", attrs=['bold']))
```
