```python
# Recorremos cada fila de la tabla, y para cada fila, recorremos cada columna (actor) y vemos si el actor está en la escena.
# Si el actor está en la escena, sumamos 1 a la cantidad de escenas en las que aparece el actor.
# Si el actor no está en la escena, no hacemos nada.

# Por lo tanto, la complejidad es O(np) donde n es la cantidad de escenas y p es la cantidad de actores

```

# Diseño
- ¿Que técnica utilizo? ¿Por qué?

```python
# Dada la dimensionalidad de los datos, utilizaré una técnica voraz para resolver el problema.
# Además, añado una poda heurística de entrada (actor con menos escenas) con la finalidad de reducir dimensiones en el espacio de solución.

# La técnica consiste en ir seleccionando el actor con menor cantidad de escenas en las que aparece.
# Luego eliminamos las escenas en las que aparece ese actor.
# De esta forma, reducimos la cantidad de escenas en las que aparece ese actor, y así sucesivamente con los demás actores.
```

# Importamos base de datos

```python
# Librerias
import pandas as pd
import re
```

```python
def convert_google_sheet_url(url):
    # Regular expression to match and capture the necessary part of the URL
    pattern = r'https://docs\.google\.com/spreadsheets/d/([a-zA-Z0-9-_]+)(/edit#gid=(\d+)|/edit.*)?'

    # Replace function to construct the new URL for CSV export
    # If gid is present in the URL, it includes it in the export URL, otherwise, it's omitted
    replacement = lambda m: f'https://docs.google.com/spreadsheets/d/{m.group(1)}/export?' + (f'gid={m.group(3)}&' if m.group(3) else '') + 'format=csv'

    # Replace using regex
    new_url = re.sub(pattern, replacement, url)

    return new_url
```

```python
# Replace with your modified URL
url = 'https://docs.google.com/spreadsheets/d/1Ipn6IrbQP4ax8zOnivdBIw2lN0JISkJG4fXndYd27U0/edit#gid=0'

new_url = convert_google_sheet_url(url)

print(new_url)
# https://docs.google.com/spreadsheets/d/1mSEJtzy5L0nuIMRlY9rYdC5s899Ptu2gdMJcIalr5pg/export?gid=1606352415&format=csv
```

```python
# Importamos el archivo
df = pd.read_csv(new_url,header=1, index_col=0)
df = df.head(30)
df = df.iloc[:,:-2]
df
```

```python
# crea una funcion que calcula el maximo de cada columna y retorna la suma de los maximos
def maximo(df):
    # agrega una fila con el maximo de cada columna
    df.loc['max'] = df.max()
    # calcula la suma de las columnas 1,2,3,4,5,6,7,8,9 y 10 donde se encuentran los maximos
    df['suma_maximos'] = df.iloc[:,0:10].sum(axis=1)
    # retorna columna suma_maximos en index max
    return df['suma_maximos']['max']
```

```python
# crea una funcion que calcula el minimo de cada columna y retorna la suma de los minimos
def greedy_optimize(_data):

    days = []
    data = _data.copy()
    while len(data) > 0:
        costs_by_actor = data.sum()
        # Toma el actor con menor cantidad de escenas
        actor = costs_by_actor.idxmin()
        # Toma las escenas en las que aparece el actor
        scenes = data[data[actor] > 0].index
        # Obtiene los actores en las escenas del actor con menor cantidad de escenas
        actors = (data.loc[scenes].sum() > 0).astype(int)
        # Toma las escenas en las que aparecen los actores seleccionados
        scenes = ((data - actors) <= 0).all(axis=1)
        # Obtiene el indice de las primeras seis escenas
        scenes = scenes[scenes].index[:6]
        days.append(scenes)
        
        # Quita las escenas consideradas
        data = data.drop(scenes)
        data = data.drop(actor, axis=1)
        
    # Elimina los dias vacios
    days = [day for day in days if len(day) > 0]
    return days
```

```python
from termcolor import colored

days = greedy_optimize(df)
maximos = []
# para cada elemento de days, imprime el numero de dia y los actores en el dia
for i, day in enumerate(days):
    print(colored(f'Day {i+1}','red', attrs=['bold']))
    # imprime los actores con max igual a 1
    print(f'Actors: {", ".join(df.loc[day].max()[df.loc[day].max() == 1].index)}')
    # imprime las escenas en el dia
    print(f'Scenes: {", ".join(day)}')
    # imprime el costo del dia
    maximos.append(maximo(df.loc[day]))
    #print(f'Cost: {df.loc[day].sum().sum()}')
    print(f'Cost: {maximo(df.loc[day])}')
    # imprime una linea en blanco
    #print()
    

print("")
print(colored(f"Total cost: {sum(maximos)}", attrs=['bold']))
```
